```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la gráfica
num_particles = 100  # Número de partículas
x_limits = (0, 80)   # Límites del eje X
y_limits = (0, 80)   # Límites del eje Y
particle_size = 30   # Tamaño de las partículas (constante para todas)

# Generar posiciones aleatorias para las partículas
x_positions = np.random.uniform(x_limits[0], x_limits[1], num_particles)
y_positions = np.random.uniform(y_limits[0], y_limits[1], num_particles)

# Crear la gráfica
plt.figure(figsize=(8, 8))
plt.scatter(x_positions, y_positions, s=particle_size, c='blue', alpha=0.7, edgecolors='black')

# Configurar límites y detalles
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Partículas en posiciones aleatorias')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la gráfica
num_particles_per_row = 15  # Número de partículas por fila
num_particles_per_col = 15  # Número de partículas por columna
particle_size = 200          # Tamaño de las partículas
spacing = 50.0               # Espaciado entre partículas

# Generar posiciones ordenadas para las partículas
x_positions = np.linspace(0, spacing * (num_particles_per_row - 1), num_particles_per_row)
y_positions = np.linspace(0, spacing * (num_particles_per_col - 1), num_particles_per_col)
x_grid, y_grid = np.meshgrid(x_positions, y_positions)

# Aplanar las matrices para usarlas en scatter
x_flat = x_grid.flatten()
y_flat = y_grid.flatten()

# Crear la gráfica
plt.figure(figsize=(8, 8))
plt.scatter(x_flat, y_flat, s=particle_size, c='blue', alpha=0.7, edgecolors='black')

# Configurar límites y detalles
plt.xlim(-spacing, spacing * num_particles_per_row)
plt.ylim(-spacing, spacing * num_particles_per_col)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Partículas ordenadas en cuadrícula')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
